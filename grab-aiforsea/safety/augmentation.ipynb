```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('merged.csv')
df.head()
```

```python
df.info()
```

```python
from sklearn.cluster import MiniBatchKMeans
```

```python
X = np.vstack((df[['acceleration_x','acceleration_y','acceleration_z']],
               df[['gyro_x','gyro_y','gyro_z']]))
```

```python
kmeans = MiniBatchKMeans(n_clusters=8**2, batch_size=32**3).fit(X)
```

```python
df['acceleration_cluster'] = kmeans.predict(df[['acceleration_x','acceleration_y','acceleration_z']])
df['gyro_cluster'] = kmeans.predict(df[['gyro_x','gyro_y','gyro_z']])
```

```python
from sklearn.decomposition import PCA

pca = PCA().fit(X)
```

```python
acceleration_pca = pca.transform(df[['acceleration_x','acceleration_y','acceleration_z']])
```

```python
df['acceleration_pca0'] = acceleration_pca[:,0]
df['acceleration_pca1'] = acceleration_pca[:,1]
df['acceleration_pca2'] = acceleration_pca[:,2]
```

```python
gyro_pca = pca.transform(df[['gyro_x','gyro_y','gyro_z']])
```

```python
df['gyro_pca0'] = gyro_pca[:,0]
df['gyro_pca1'] = gyro_pca[:,1]
df['gyro_pca2'] = gyro_pca[:,2]
```

```python
df.head()
```

```python
df.to_csv('augmentation.csv', index = False)
```
